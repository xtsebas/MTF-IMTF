## a) Calcular el costo de acceso utilizando el algoritmo MTF para

Imprima la lista de configuración, la solicitud, su costo y la configuración de la lista aplicando MTF
por cada solicitud en la secuencia y, al final, imprima el costo total de los accesos.


```python
config = [0, 1, 2, 3, 4]
requests = [0, 1, 2, 3, 4] * 4

total, history = move_to_front(config.copy(), requests)

for step in history:
    print(f"Lista Antes: {step['before']},        Solicitud: {step['request']},       Costo: {step['cost']},     Lista Después: {step['after']}")
print("Costo total:", total)
```

## e) Calcular el costo de acceso utilizando el algoritmo MTF

```python
from algorithms.MFT import move_to_front
from algorithms.IMFT import improved_move_to_front
import pandas as pd
config = [0, 1, 2, 3, 4]
```

### Secuencias Repetidas de 2 y 3 

```python
seq_2 = [2]*20
seq_3 = [3]*20

cost_e2, hist_e2 = move_to_front(config.copy(), seq_2.copy())
cost_e3, hist_e3 = move_to_front(config.copy(), seq_3.copy())

df_e = pd.DataFrame({
    'Iteración': list(range(1,21))*2,
    'Solicitud': ['2']*20 + ['3']*20,
    'Costo': [h['cost'] for h in hist_e2] + [h['cost'] for h in hist_e3]
})
df_e
```

### Comparativa MTF vs IMTF

```python
seq_c = [0]*20
seq_d = [4,3,2,1,0]*4

mtf_c, _  = move_to_front(config.copy(), seq_c.copy())
imtf_c, _ = improved_move_to_front(config.copy(), seq_c.copy())
mtf_d, _  = move_to_front(config.copy(), seq_d.copy())
imtf_d, _ = improved_move_to_front(config.copy(), seq_d.copy())

df_f = pd.DataFrame([
    {'Secuencia':'C (Mejor caso)','MTF':mtf_c,'IMTF':imtf_c},
    {'Secuencia':'D (Peor caso)','MTF':mtf_d,'IMTF':imtf_d},
])
df_f
```

```python
import matplotlib.pyplot as plt

# Gráfico de df_e
for label, grp in df_e.groupby('Solicitud'):
    plt.plot(grp['Iteración'], grp['Costo'], marker='o', label=f'Elemento {label}')
plt.xlabel('Iteración')
plt.ylabel('Costo')
plt.legend()
plt.title('Costos por Iteración – Inciso e')
```
